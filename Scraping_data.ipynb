# Scraping climate data by month and creating a master database

```python
import requests
import lxml.html as lh
import pandas as pd
import numpy as np
import itertools as it
```

### January

```python
def jan(j):
    page = requests.get('https://en.tutiempo.net/climate/01-' + str(j)+ str('/ws-484550.html'))
    doc = lh.fromstring(page.content)
    tr_elements = doc.xpath('//tr')
    col=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
    T=tr_elements[36]
    i=0
    for t in T.iterchildren():
        data=t.text_content() 
        if i>0:
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i].append(data)
        #Increment i for the next column
        i+=1    
    return(col)  

#Making dataframe
df_jan= pd.DataFrame()
for j in it.chain(range(1954,1966), range(1973,1995),range(1996,2020)) :
    df_jan=df_jan.append(pd.DataFrame(jan(j)).T)
    
list_year = [*range(1954,1966,1),*range(1973,1995,1),*range(1996,2020,1)]
df_jan["Year"]=list_year

df_jan.columns=["index","avgT","maxT","minT","slp","avgRH","PP","VV","avgW","VM","maxW","RA","SN","TS","FG","Year"]
df_jan.reset_index(drop=True, inplace=True)
df_jan['index']= 1
```

```python
def feb(j):
    page = requests.get('https://en.tutiempo.net/climate/02-' + str(j)+ str('/ws-484550.html'))
    doc = lh.fromstring(page.content)
    tr_elements = doc.xpath('//tr')
    col=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
    T=tr_elements[33]
    i=0
    for t in T.iterchildren():
        data=t.text_content() 
        if i>0:
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i].append(data)
        #Increment i for the next column
        i+=1    
    return(col)  

#Making dataframe
df_feb= pd.DataFrame()
for j in it.chain(range(1954,1966), range(1973,1995),range(1996,2020)) :
    df_feb=df_feb.append(pd.DataFrame(feb(j)).T)
    
list_year = [*range(1954,1966,1),*range(1973,1995,1),*range(1996,2020,1)]
df_feb["Year"]=list_year

df_feb.columns=["index","avgT","maxT","minT","slp","avgRH","PP","VV","avgW","VM","maxW","RA","SN","TS","FG","Year"]
df_feb.reset_index(drop=True, inplace=True)



def feb_leap(j):
    page = requests.get('https://en.tutiempo.net/climate/02-' + str(j)+ str('/ws-484550.html'))
    doc = lh.fromstring(page.content)
    tr_elements = doc.xpath('//tr')
    col=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
    T=tr_elements[34]
    i=0
    for t in T.iterchildren():
        data=t.text_content() 
        if i>0:
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i].append(data)
        #Increment i for the next column
        i+=1    
    return(col) 

df_feb_leap = pd.DataFrame()
for j in [1956,1960,1964,1976,1980,1984,1988,1992,1996,2000,2004,2008,2012,2016] :
    df_feb_leap= df_feb_leap.append(pd.DataFrame(feb_leap(j)).T)

list_year = [1956,1960,1964,1976,1980,1984,1988,1992,1996,2000,2004,2008,2012,2016]
df_feb_leap["Year"]=list_year

df_feb_leap.columns=["index","avgT","maxT","minT","slp","avgRH","PP","VV","avgW","VM","maxW","RA","SN","TS","FG","Year"]
df_feb_leap.reset_index(drop=True, inplace=True)


# Replacing rows in feb dataframe with feb_leap year dataframe
df_feb.loc[2]=df_feb_leap.loc[0]
df_feb.loc[6]=df_feb_leap.loc[1]
df_feb.loc[10]=df_feb_leap.loc[2]
df_feb.loc[15]=df_feb_leap.loc[3]
df_feb.loc[19]=df_feb_leap.loc[4]
df_feb.loc[23]=df_feb_leap.loc[5]
df_feb.loc[27]=df_feb_leap.loc[6]
df_feb.loc[31]=df_feb_leap.loc[7]
df_feb.loc[34]=df_feb_leap.loc[8]
df_feb.loc[38]=df_feb_leap.loc[9]
df_feb.loc[42]=df_feb_leap.loc[10]
df_feb.loc[46]=df_feb_leap.loc[11]
df_feb.loc[50]=df_feb_leap.loc[12]
df_feb.loc[54]=df_feb_leap.loc[13]

df_feb['index']= 2
```

### Combining dataframes

```python
frames= [df_jan, df_feb, df_mar, df_apr, df_may, df_jun, df_jul, df_aug, df_sep, df_oct, df_nov, df_dec]
df= pd.concat(frames)
```

### Sorting the dataframe

```python
df_c=df.sort_values(by=['Year','index'],ascending=True)
```

```python
df_c = df_c[['index','Year','avgT','maxT','minT','slp','avgRH','PP','VV','avgW','VM','maxW','RA','SN','TS','FG']]
```
